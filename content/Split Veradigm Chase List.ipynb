```python
file_path = r"C:\Users\aveesh.kachroo\OneDrive - datalinksoftware.com\Desktop\Chase Lists\VERADIGM\CENTENE\CenteneChaselist20240214.xlsx"
og_chase_list = pd.read_excel(file_path, dtype = {'Zip': str})
og_chase_list
```
